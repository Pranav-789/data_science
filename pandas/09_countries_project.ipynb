```python
## country with highest population
df[df['population'] == df['population'].max()]['country']
```

```python
## capital of country with the highest population
df[df['population'] == df['population'].max()]['capital_city']
```

```python
## country with least population
df[df['population'] == df['population'].min()]['country']
```

```python
## capital of country with the least population
df[df['population'] == df['population'].min()]['capital_city']
```

```python
## top 5 countries with highest democratics score
## democracy_score

df.sort_values(by = 'democracy_score', ascending=False, inplace=True)
```

```python
df.head()
```

```python
# Count how many times each region appears in the dataset
# (i.e., frequency of each unique region)
df['region'].value_counts()


# Get the number of unique regions
# value_counts() gives frequencies per region
# count() counts how many unique regions exist
total_regions = df['region'].value_counts().count()
```

```python
## how many countries lie in Eastern Europe region
# df[df['region']=='Eastern Europe'].shape[0]
## or
df['region'].value_counts()['Eastern Europe']
```

```python
df[df['region'] == 'Eastern Europe']['country']
```

```python
## political leader of second highest populated couuntry

df[df['population'] == df['population'].nlargest(2).iloc[1]]['political_leader'] ## iloc is is indied from 0 to n-1
```

```python
## how many countries are there whose political leaders are unknown
df[df['political_leader'].isna()]['country'].count()
```

```python
## how many countries have republic in their full name

df[df['country_long'].apply(lambda x: 'republic' in str(x).lower())]
```

```python
## which country in african region has the highest population

african_df = df[df['continent'] == 'Africa']
```

```python
african_df[african_df['population'] == african_df['population'].max()]
```
