```python
## make a new column for episode count
def extract_ep(st: str):
    if '(' not in st or ')' not in st:
        return None
    
    data = st[st.find('(')+1: st.find(')')]
    return data

df1['episode_cnt'] = df1['Title'].apply(extract_ep)
```

```python
df1.head()
df1['episode_cnt'] = df1['episode_cnt'].apply(lambda x : x.replace(" eps", ""))
```

```python
df1.head()
```

```python
df1['episode_cnt'] = df1['episode_cnt'].astype(int)
df1.head()
```

```python
df1.loc[0]['Title']
```

```python
def extraction_time(txt):
    return txt[txt.find(')')+1 : txt.find(')')+20]

print(extraction_time(df1.loc[0]['Title']))
```

```python
df1['timestamps'] = df1['Title'].apply(extraction_time)
```

```python
df1.head()
```

```python
from dateutil.relativedelta import relativedelta
from datetime import datetime

def calculate_total_months(period):
    try:
        start_str, end_str = period.split(' - ')
        start_date = datetime.strptime(start_str, '%b %Y')
        end_date = datetime.strptime(end_str, '%b %Y')
        r = relativedelta(end_date, start_date)
        return r.years * 12 + r.months + 1  # +1 to include the starting month
    except:
        return None

df1['Months'] = df1['timestamps'].apply(calculate_total_months)
```
