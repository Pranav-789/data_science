```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# -----------------------------
# REGRESSION PLOT (LM PLOT)
# -----------------------------

# lmplot() is used to visualize the LINEAR RELATIONSHIP
# between two numerical variables using a regression line.

sns.lmplot(
    x='total_bill',   # Independent variable (predictor) on the x-axis
    y='tip',          # Dependent variable (response) on the y-axis
    data=tips,        # DataFrame containing the variables
    hue='sex'         # Categorical variable used to create separate regressions
)

# This plot shows:
# 1. Scatter points representing actual data
# 2. A best-fit regression line for each 'sex' category
# 3. Confidence interval around each regression line
```
