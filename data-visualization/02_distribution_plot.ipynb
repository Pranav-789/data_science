```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## seaborn has some built in datasets
## we will use tips.csv
```

```python
df = sns.load_dataset('tips') ## requires network
```

```python
## quantitve data
# total_bill
# tip

## qualitative/categorical data
# sex
# smoker
# data
# time
# size

df['size'].unique()
## out of 244 only uniques values are array([2, 3, 4, 1, 6, 5])
## therefore size is also a categorical data
```

```python
# -----------------------------
# DISTRIBUTION PLOT
# -----------------------------

# histplot() is used to visualize the distribution of a numerical variable.
# Here, we are plotting the distribution of the 'total_bill' column
# from the DataFrame 'df'.
plt.subplot(1, 2, 1)
sns.histplot(
    df['total_bill'],   # Numerical data whose distribution we want to analyze
    bins=20,             # Number of bars (intervals) the data range is divided into
    kde=True             # Overlay a KDE (smooth probability curve) on the histogram
)
plt.subplot(1, 2, 2)
sns.histplot(
    df['tip'],   # Numerical data whose distribution we want to analyze
    bins=20,             # Number of bars (intervals) the data range is divided into
    kde=True             # Overlay a KDE (smooth probability curve) on the histogram
)
# This creates:
# 1. A histogram → shows frequency of values
# 2. A KDE curve → shows the smooth distribution shape
```

```python
# -----------------------------
# JOINT PLOT
# -----------------------------

# jointplot() is used to visualize the relationship between TWO numerical variables
# at the same time, along with their individual distributions.

sns.jointplot(
    x='total_bill',   # Variable plotted on the x-axis (independent variable)
    y='tip',          # Variable plotted on the y-axis (dependent variable)
    data=df,          # DataFrame that contains both variables
    kind='scatter'    # Type of joint plot: scatter shows individual data points
)

# This creates:
# 1. A scatter plot in the center (relationship between total_bill and tip)
# 2. A histogram/KDE on the top (distribution of total_bill)
# 3. A histogram/KDE on the right (distribution of tip)
```

```python
sns.jointplot(
    x='total_bill',   # Variable plotted on the x-axis (independent variable)
    y='tip',          # Variable plotted on the y-axis (dependent variable)
    data=df,          # DataFrame that contains both variables
    kind='hex'    # Type of joint plot: hex
)
```

```python
sns.jointplot(
    x='total_bill',   # Variable plotted on the x-axis (independent variable)
    y='tip',          # Variable plotted on the y-axis (dependent variable)
    data=df,          # DataFrame that contains both variables
    kind='kde'    # Type of joint plot: kde
)
```

```python
sns.jointplot(
    x='total_bill',   # Variable plotted on the x-axis (independent variable)
    y='tip',          # Variable plotted on the y-axis (dependent variable)
    data=df,          # DataFrame that contains both variables
    kind='reg'    # Type of joint plot: regression
)
```

```python
# -----------------------------
# PAIR PLOT
# -----------------------------

# pairplot() creates a matrix of plots to visualize
# relationships between MULTIPLE numerical variables at once.

sns.pairplot(
    df,                 # DataFrame containing the variables to analyze
    hue='size',         # Categorical column used to color-code the data points
    palette='rainbow'   # Color palette applied to different 'size' categories
)

# This generates:
# 1. Scatter plots for every pair of numerical features
# 2. Distribution plots (hist/KDE) on the diagonal
# 3. Color separation based on the 'size' category
```
