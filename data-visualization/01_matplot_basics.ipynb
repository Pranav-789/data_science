## Intro to Matplotlib

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 11)
x
```

```python
y = x**3
y
```

```python
plt.plot(x, y)
plt.title('Hello')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

## Subplots in matplotlib

```python
### one canvas having abouts r*c cells
## r -> rows
## c -> columns

## plot 1
plt.subplot(2, 2, 1)
plt.title('1st plot')
plt.plot(x, y)

## plot 2
plt.subplot(2, 2, 2)
plt.title('2nd plot')
plt.plot(x**2, x**2)

## plot 3
plt.subplot(2, 2, 3)
plt.title('3rd plot')
plt.plot(y, x)

## plot 4
plt.subplot(2, 2, 4)
plt.plot(x, y)
plt.title('4th plot')
```

### Objet oriented way

```python
# Step 1: Create a Figure object
# --------------------------------
# A Figure represents the ENTIRE drawing canvas.
# Think of it as a blank sheet of paper or a window.
# At this point, the figure exists but has NO plots (Axes) inside it.
# fig = plt.figure()  
# Output: <Figure size 640x480 with 0 Axes>
fig = plt.figure(figsize=(8, 8), dpi=100)  
# figsize controls how big the canvas is,
# dpi controls how sharp it is

# Step 2: Add an Axes object to the Figure
# ----------------------------------------
# An Axes is the actual plotting area where data is drawn.
# A Figure can contain multiple Axes objects.

# fig.add_axes() takes a list of 4 values:
# [left, bottom, width, height]
# - All values are in RELATIVE units (0 to 1)
# - Relative to the Figure's width and height

axis1 = fig.add_axes([0.1, 0.1, 1, 1])

# Meaning:
# left   = 0.1  -> start 10% from the left edge of the figure
# bottom = 0.1  -> start 10% from the bottom of the figure
# width  = 1.0  -> take 100% of the figure width
# height = 1.0  -> take 100% of the figure height


# Step 3: Plot data on the first Axes
# -----------------------------------
# axis1.plot() draws the line on THIS Axes only.
# This is the key idea of object-oriented plotting:
# "The Axes owns the plot, not pyplot."
axis1.plot(x, y)


# Step 4: Add another Axes on the SAME Figure
# -------------------------------------------
# This creates a second plotting area on top of the first one.
# Useful for insets, zoomed views, or comparisons.

axis2 = fig.add_axes([0.3, 0.4, 0.5, 0.5])

# Meaning:
# left   = 0.3  -> start 30% from the left of the figure
# bottom = 0.4  -> start 40% from the bottom
# width  = 0.5  -> occupy 50% of the figure width
# height = 0.5  -> occupy 50% of the figure height


# Step 5: Plot data on the second Axes
# -----------------------------------
# This plot is independent of axis1.
# Each Axes maintains its own limits, ticks, labels, and data.
axis2.plot(x, y)
```

### working with images
