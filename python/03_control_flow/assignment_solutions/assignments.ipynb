```python
num = int(input("Enter a number: "))
if num == 1 or num == 2:
    print("Not prime!")
from math import sqrt
for i in range(2, int(sqrt(num))+1):
    if(num%i==0):
        print("Not prime!")
        break
else:
    print("Prime!")
```
