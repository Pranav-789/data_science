```python
### Read a whole file

with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
## Read a file line by line

with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip()) #.strip removes the newline character
```

```python
## Writing a file(Overwriting)

with open('example.txt', 'w') as file:
    file.write("Hello world!\n")
    file.write("This is a new line.")
```

```python
## Write a file without overwriting

with open('example.txt', 'a') as file:
    file.write(" Append operation taking place.")
```

```python
## Writing list of lines to a file

lines = ['First line \n', 'Second line \n', 'Third line \n']

with open('example.txt', 'a') as file:
    file.writelines(lines)
```

```python
### Binary files

data = b'\x00\x01\x02\x03\x04'
with open('ex.bin', 'wb') as file:
    file.write(data)
```

```python
with open('ex.bin', 'rb') as file:
    content = file.read()
    print(content)
```

```python
### Read the content from src and write it in destination file
with open('example.txt', 'r') as file:
    content = file.read()

with open('destination.txt', 'w') as file:
    file.write(content)
```

```python
### Writing and then reading a file

with open('example.txt', 'w+') as file:
    file.write("Hello World\n")
    file.write("This is a new line\n")

    ## Move the file cursor to the beginning
    file.seek(0)
    content = file.read()
    print(content)
```

```python
## create a new directory
new_directory = 'package'
import os
os.mkdir(new_directory)
print(f'Directory {new_directory} is created')
```

```python
## Listing files and directories

items = os.listdir('.')
print(items)
```

```python
## Joining Paths

fdir_name = 'folder'
file_name='file.txt'

full_path = os.path.join(fdir_name, file_name)
print(full_path)
```

```python
## Joining Paths

fdir_name = 'folder'
file_name='file.txt'

full_path = os.path.join(os.getcwd(), fdir_name, file_name)
print(full_path)
```

```python
path = 'example1.txt'

if(os.path.exists(path)):
    print(f'the path {path} exists')
else:
    print(f'the path {path} does not exists')
```

```python
## Checking if a path is a file or a directory
path = 'example.txt'

if os.path.isfile(path):
    print(f"The path {path} is a file")
elif os.path.isdir(path):
    print(f"The path {path} is a dir")
else:
    print(f"The path {path} is not a file or a directory")
```

```python
### getting the absolute path
relative_path = 'example.txt'
absolute_path = os.path.abspath(relative_path)
print(absolute_path)
```
