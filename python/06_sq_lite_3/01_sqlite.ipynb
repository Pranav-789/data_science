```python
import sqlite3
```

```python
connection = sqlite3.connect('example.db')
connection
```

```python
cursor = connection.cursor()
```

```python
## Create a table
cursor.execute('''
Create Table If Not Exists employees(
    id Integer Primary Key,
    name Text Not Null,
    age Integer,
    department text  
    )
''')

## commit the changes
connection.commit()
```

```python
cursor.execute('''
Select * from employees
''')
```

```python
## Insert data in sqlite3
cursor.execute('''
Insert into employees(name, age, department) values('Krish', 32, 'Data Scientist')
''')

cursor.execute('''
Insert into employees(name, age, department) values('Charlie', 35, 'Finance')
''')

connection.commit()
```

```python
## Query the data from the table
cursor.execute('''
Select * from employees
''')

rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
## Update the data in the table

cursor.execute('''
Update employees
set age = 20
where name = 'Krish'
''')

connection.commit()
```

```python
## Query the data from the table
cursor.execute('''
Select * from employees
''')

rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
## Delete data from table
cursor.execute('''
Delete from employees
where id in (3, 4)
''')

connection.commit()
```

```python
## Query the data from the table
cursor.execute('''
Select * from employees
''')

rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
## Working with sales data

connection = sqlite3.connect('sales_data.db')
cursor = connection.cursor()

cursor.execute('''
Create Table if not exists sales(
    id INTEGER PRIMARY KEY,
    date TEXT NOT NULL,
    product TEXT NOT NULL,
    sales INTEGER,
    region TEXT               
)
''')

sales_data = [
    {"date": "2026-01-01", "product": "Laptop", "sales": 10, "region": "North"},
    {"date": "2026-01-01", "product": "Mouse", "sales": 50, "region": "South"},
    {"date": "2026-01-02", "product": "Keyboard", "sales": 30, "region": "East"},
    {"date": "2026-01-02", "product": "Monitor", "sales": 20, "region": "West"},
    {"date": "2026-01-03", "product": "Laptop", "sales": 15, "region": "North"},
    {"date": "2026-01-03", "product": "Mouse", "sales": 40, "region": "South"}
]

connection.commit()
```

```python
cursor.executemany('''
    INSERT INTO sales(date, product, sales, region)
    VALUES (:date, :product, :sales, :region)
''', sales_data)

connection.commit()
```

```python
## Query the data from the table
cursor.execute('''
Select * from sales
''')

rows = cursor.fetchall()

for row in rows:
    print(row)
```
