```python
import os
dir_list = set(os.listdir())
os.mkdir('new_dir')
new_list = set(os.listdir())

for new in (new_list - dir_list):
    if os.path.isfile(new):
        os.remove(new)
    elif os.path.isdir(new):
        os.rmdir(new)
```
