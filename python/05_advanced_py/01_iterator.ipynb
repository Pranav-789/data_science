```python
## generators are subclass of iterators
def square(n):
    for i in range(3):
        yield i**2

gen = square(3)
print(next(gen))
print(next(gen))
print(next(gen))
# print(next(gen)) # stop iteration error
```

```python
gen = square(4)
print(gen)
```

```python
for val in gen:
    print(val)
```

```python
### Practical : Reading large files

def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line
```

```python
gen = read_large_file('large_article.txt')

for line in gen:
    print(line.strip(), end='\n\n')
```
