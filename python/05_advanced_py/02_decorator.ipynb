```python
## closures

def main_welcome(msg):
    def sub_welcome_method():
        print("Welcome to advanced python")
        print(msg)
    return sub_welcome_method
```

```python
sub = main_welcome(0)
sub()
```

```python
def main_welcome(func):
    def sub_welcome_method():
        print("Welcome to advanced python")
        func("welcome everyone")
    return sub_welcome_method
```

```python
sub = main_welcome(print)
sub()
```

```python
### Decorator
def main_welcome(func):
    def sub_welcome_method():
        print("Welcome to advanced python")
        func()
    return sub_welcome_method()
```

```python
def course_introduction():
    print("This is an advanced python course")

course_introduction()
```

```python
@main_welcome
def course_introduction(): ## this function go to main_welcome as a parameter
    print("This is an advanced python course")
```

```python
## Decorator


def my_decorator(func):
    def wrapper():
        print("Something is happening before the func is called")
        func()
        print("Something is happening after the func is called")
    return wrapper
```

```python
@my_decorator
def say_hello():
    print("Hello!")
```

```python
say_hello()
```

```python
## decorators with arguments
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for i in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print("hello")
```

```python
say_hello()
```
